```python
# magical line to print image in line
%matplotlib inline 

# importing necessary files
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# opening image
path = '../images/gal.jpg'
img = cv.imread(path,cv.IMREAD_COLOR)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
img_r = np.zeros(img.shape,dtype=np.uint8)
img_r[:,:,0]=img[:,:,0]

img_gray = cv.add(img_gray,100)

# display image
f = plt.figure(figsize=(16,4))
ax1 = f.add_subplot(151)
ax2 = f.add_subplot(155)
ax3 = f.add_subplot(152)
ax1.imshow(img)
ax1.title.set_text("original")
ax1.set_xticks([])
ax1.set_yticks([])
ax2.imshow(img_gray,cmap='gray')
ax2.title.set_text("grayscale")
ax2.set_xticks([])
ax2.set_yticks([])
ax3.imshow(img_r)
ax3.title.set_text("red")
ax3.set_xticks([])
ax3.set_yticks([])
plt.show()

print(img_gray.shape)
```
