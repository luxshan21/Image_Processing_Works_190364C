```python
# Question 01

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = cv.imread('sudoku.png',cv.IMREAD_COLOR)
assert im is not None
gray = cv.cvtColor(im,cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray, 20, 120, apertureSize=3)
lines = cv.HoughLines(edges,1,np.pi/180,175)

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0, y0 = a*rho, b*rho
    x1 ,y1 = int(x0+1000*(-b)), int(y0+1000*(a))
    x2 ,y2 = int(x0-1000*(-b)), int(y0-1000*(a))
    cv.line(im,(x1,y1),(x2,y2),(0,0,255),2)
fig, axes = plt.subplots(1,3, figsize=(16,16))
axes[0].imshow(cv.cvtColor(gray, cv.COLOR_BGR2RGB))
axes[0].set_title('Gray Image')
axes[1].imshow(cv.cvtColor(edges, cv.COLOR_BGR2RGB))
axes[1].set_title('Edges ')
axes[2].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
axes[2].set_title('Image')
for i in range(3):
 axes[i].set_xticks([]), axes[i].set_yticks([])
plt.show()
```

```python
# Question 02

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('coins.jpg',0)
img = cv.medianBlur(img,5)
cimg = cv.cvtColor(img,cv.COLOR_GRAY2BGR)
circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,20,param1=150,param2=20,minRadius=20,maxRadius=50)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
plt.axis('off')
plt.title('Detected Circles')
plt.imshow(cv.cvtColor(cimg, cv.COLOR_BGR2RGB), cmap = 'gray')
plt.show()
```

```python
# Question 03

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('pic1.png', cv.IMREAD_REDUCED_GRAYSCALE_2)
assert img is not None
templ = cv.imread('templ.png', cv.IMREAD_REDUCED_GRAYSCALE_2)
assert templ is not None

im_edges = cv.Canny(img, 50, 250)
templ_edges = cv. Canny(templ, 50, 250)
alg = cv.createGeneralizedHoughGuil()
alg.setTemplate(templ_edges)
alg.setAngleThresh(100000)
alg.setScaleThresh(40000)
alg.setPosThresh(1000)
alg.setAngleStep(1)
alg.setScaleStep(0.1)
alg.setMinScale(0.9)
alg.setMaxScale(1.1)
positions, votes = alg.detect(im_edges)

out = cv.cvtColor(img, cv.COLOR_BAYER_BG2BGR)
for x, y, scale, orientation in positions[0]:
    halfHeight = templ.shape[0]/2.*scale
    halfWidth = templ.shape[1]/2.*scale
    p1 = (int(x - halfWidth), int(y - halfHeight))
    p2 = (int(x + halfWidth), int(y + halfHeight))
    print("x = {}, y = {}, scale = {}, orientation = {}, p1 = {}, p2 = {}".format(x,y, scale, orientation, p1, p2))
    cv.rectangle(out, p1, p2, (0,0,255))
fig, axes = plt.subplots(1,2, sharex='all', sharey='all', figsize=(8,8))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original'),axes[0].set_xticks([]), axes[0].set_yticks([])
axes[1].imshow(out, cmap='gray')
axes[1].set_title('Out'),axes[1].set_xticks([]), axes[1].set_yticks([])
plt.show()
```

```python
# Question 05

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im1 = cv.imread('img1.ppm',cv.IMREAD_ANYCOLOR)
im4 = cv.imread ('img4.ppm',cv.IMREAD_ANYCOLOR)
H = np.array([[6.6378505e-01,6.8003334e-01, -3.1230335e+01],[-1.4495500e-01, 9.7128304e-01, 1.4877420e+02],[4.2518504e-04, -1.3930359e-05,1.0000000e+00]])
im1to4 = cv.warpPerspective(im4, np.linalg.inv(H), (2000, 2000))

fig, axes  = plt.subplots(1,3, figsize=(15,15))
axes[0].imshow(cv.cvtColor(im1, cv.COLOR_BGR2RGB), cmap='gray')
axes[0].set_title('Image 1')
axes[0].set_xticks([])
axes[0].set_yticks([])
axes[1].imshow(cv.cvtColor(im4, cv.COLOR_BGR2RGB), cmap='gray')
axes[1].set_title('Image 2')
axes[1].set_xticks([]), axes[1].set_yticks([])
axes[2].imshow(cv.cvtColor(im1to4, cv.COLOR_BGR2RGB), cmap='gray')
axes[2].set_title('Image 1 Warped')
axes[2].set_xticks([]), axes[2].set_yticks([])
plt.show()
```
