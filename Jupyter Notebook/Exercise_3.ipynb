```python
# Question 1
#  Box filer
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np


image = cv.imread('butterfly.jpg', cv.IMREAD_REDUCED_GRAYSCALE_4)
assert image is not None
box = 1./81.*np.ones((9,9))

image_avg = cv.filter2D(image,-1,box)
image_gau = cv.sepFilter2D(image, -1, cv.getGaussianKernel(9, 4), cv.getGaussianKernel(9, 4))

fig, axes  = plt.subplots(1,3, sharex='all', sharey='all', figsize=(15,15))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')
axes[0].set_xticks([]), axes[0].set_yticks([])
axes[1].imshow(image_avg, cmap='gray')
axes[1].set_title('Avg')
axes[1].set_xticks([]), axes[1].set_yticks([])
axes[2].imshow(image_gau, cmap='gray')
axes[2].set_title('Gaussian')
axes[2].set_xticks([]), axes[1].set_yticks([])
plt.show()
```

```python
# Question 3 (a) (b) (c)
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('contact_lens.tif', cv.IMREAD_GRAYSCALE).astype(np. float32)


sobel_v = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=np.float32) 
f_x = cv.filter2D(f, -1, sobel_v) 
sobel_h = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=np.float32)
f_y = cv.filter2D(f, -1, sobel_h) 
grad_mag = np.sqrt(f_x**2 + f_y**2)
                   
fig, ax = plt.subplots (1,4, figsize=(18,6)) 
ax[0].imshow(f, cmap='gray', vmin=0, vmax=255) 
ax[0].set_title('Original')
ax[1].imshow(f_x, cmap='gray', vmin=-1020, vmax=1020)
ax[1].set_title(r'Sobel Vertical $f_x$')
ax[2].imshow(f_y, cmap="gray", vmin=-1020, vmax=1020) 
ax[2].set_title(r'Sobel Horizontal$f_y$')
ax[3].imshow(grad_mag, cmap="gray")
ax[3].set_title(r'Gradient Magnitude $\sqrt{f_x^2 + f_y^2}$')

for i in range (4):
    ax[i].axis('off')
plt.show()
```

```python
# Question 4 
# Sharpenign Using  Filer

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread (r'tom.jpg', cv.IMREAD_GRAYSCALE).astype(np.float32)
assert f is not None

sigma = 2

gaussian_ld = cv.getGaussianKernel (5, sigma)
f_lp = cv.sepFilter2D (f, -1, gaussian_ld, gaussian_ld)
f_hp = f-f_lp
f_sharpened = cv.addWeighted(f, 1.0, f_hp, 2.0, 0)

fig, ax = plt.subplots (1,4, figsize=(18,6))
ax[0].imshow(f, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow (f_lp, cmap='gray')
ax[1].set_title(r'f_{lp}')
ax [2].imshow (f_hp,cmap='gray')
ax[2].set_title(r'f_{hp}')
ax[3].imshow(f_sharpened, cmap="gray")
ax[3].set_title(r'Sharpened')

for i in range (4):
    ax[i].axis ('off')
plt.show()
```
