```python
# Morphological operations

import cv2 as cv
import numpy as np

image = cv.imread("image.jpg")


kernel = np.ones((5,5), dtype = 'uint8')
#erosion = cv.erode(image, kernel, iterations = 1)
#dilation = cv.dilate(image, kernel, iterations = 5)
opening = cv.morphologyEx(image, cv.MORPH_OPEN, kernel) #erosion followed by dilation
closing = cv.morphologyEx(image, cv.MORPH_CLOSE, kernel) #dilation followed by erosion
cv.imshow("Original image ",image) #Display Original Image
#cv.imshow("erosion ",erosion) #Display after erosion image
#cv.imshow("dilation ",dilation) #Display after dilation image
cv.imshow("Opening ",opening) #Display after dilation image
cv.imshow("Closing ",closing) #Display after dilation image
cv.waitKey(0)
cv.destroyAllWindows()
0
```

```python
#Image Transformation

import cv2 as cv
import numpy as np

img = cv.imread("girl1.jpg")
cols= img.shape[1]
rows = img.shape[0]

M = np.float32([(1,0,150), (0,1,70)])
shifted = cv.warpAffine (img, M, (cols, rows))
cv.imshow('original image', img) 
cv.imshow('shift image', shifted)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#rotation of the image

import cv2 as cv
img=cv.imread("girl1.jpg")

rows=img.shape[1]
cols=img.shape[0]

centre = (cols/2,rows/2)
angle = 45

M = cv.getRotationMatrix2D(centre, angle, 1) 
rotate = cv.warpAffine (img, M, (cols,rows))

cv.imshow('original image', img) 
cv.imshow('rotated', rotate) 
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Continue rotation of Image

import cv2 as cv
img=cv.imread("girl1.jpg")

rows=img.shape[1]
cols=img.shape[0]

centre = (cols/2,rows/2)
for angle in range(10,180,5):
    M = cv.getRotationMatrix2D(centre, angle, 1) 
    rotate = cv.warpAffine (img, M, (cols,rows))
    cv.imshow('rotated', rotate) 
    cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Gaussian Blur 
# to remove image Noices

import cv2 as cv

pic = cv.imread("gal_gaussian.png") 
resize = cv.resize(pic, (640,640))
ksize=(7,7)
sigmax=0
# sigmay = None

blur= cv.GaussianBlur(pic, ksize,sigmax) 
resizeblur= cv.resize(blur, (640,640))

cv.imshow("Gaussian Input", resize)
cv.imshow('Gaussian Output', resizeblur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Median Blur

import cv2 as cv

pic= cv.imread("gal_gaussian.png") 
resize = cv.resize(pic, (520,520))

kernal = 3
sigmax = 0

mblur= cv.medianBlur(pic, kernal) 
resizeblur= cv.resize(mblur, (520,520))

cv.imshow('Median Input', resize) 
cv.imshow('Median Output', resizeblur)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
# Bilateral Filter

import cv2 as cv

pic= cv.imread("gal_gaussian.png")

dimpixel = 7
sigmacolor= 100 
sigmaspace = 100
bilate = cv.bilateralFilter(pic,dimpixel,sigmacolor,sigmaspace)

cv.imshow('input',pic) 
cv.imshow("Output",bilate)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Face Detection Using Image

import cv2 as cv

face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')

pic = cv.imread("girl1.jpg")
scale_factor = 1.3

while(True): 
    faces = face_cascade.detectMultiScale(pic, scale_factor, 5)
    for(x,y,w,h) in faces:
        cv.rectangle(pic, (x,y), (x+w, y+h), (255,0,0), 2) 
        font = cv.FONT_HERSHEY_SIMPLEX 
    cv.putText(pic, 'girl', (x,y), font, 2, (255,255,0), 2, cv.LINE_AA)

    print('No of face found', format(len(faces))) 
    cv.imshow('face', pic) 
    k = ( cv.waitKey(30) & 0xff )
    if k == 2: 
        break

cv.destroyAllWindows()
```
