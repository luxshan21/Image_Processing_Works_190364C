```python
# Python program for implementation of Insertion Sort
import time
import matplotlib.pyplot as plt

# Function to do insertion sort
def insertionSort(arr):

	# Traverse through 1 to len(arr)
	for i in range(1, len(arr)):

		key = arr[i]

		# Move elements of arr[0..i-1], that are
		# greater than key, to one position ahead
		# of their current position
		j = i-1
		while j >= 0 and key < arr[j] :
				arr[j + 1] = arr[j]
				j -= 1
		arr[j + 1] = key


# Driver code to test above
arr = [12, 11, 13, 5, 6]
insertionSort(arr)
for i in range(len(arr)):
	print ("% d" % arr[i])

import random


times = []
list2 = []
n=10
for x in range(n):
  r = random.randint(20,100)
  list2.append(r)

  for y in range(10):
    t+=exe_time_eval(x)
  times.append((t/10))
print(times)
import matplotlib.pyplot as plt
elements = range(n)
plt.xlabel('List Length')
plt.ylabel('Time Complexity')
plt.plot(elements, times, label ='loop')
plt.grid()
plt.legend()
plt.show()
```

```python
P1 = K1 @ np.hstack((R1, t1)) # P = K*[R/t]
P2 = K2 @ np.hstack ((R2, t2)) # P = K*[R[t]


from scipy. linalg import null_space

def skew(x):
    x = x.ravel()
    return np.array([[0, -x[2], x[1]], [x[2], 0, -x[0]], [-x[1], x[0], 0]])

C = null_space(P1)
C = C * np.sign(C[0,0])
e2 = P2 @ C
e2x = skew(e2)
F = e2x @ P2 @ np.linalg.pinv(P1)
print('F=', F)
```

```python
def drawlines(img1,lines, pts):
    r,c,_ = img1.shape
    for r, pt1 in zip(lines, pts):
        color=(0,255,255)
        x0,y0 = map(int, [0, -r[2]/r[1] ])
        x1,y1 = map(int, [c, -(r[2]+r[0]*c)/r[1] ])
        img1 = cv.line(img1, (x0,y0), (x1,y1), color ,4)
        img1 = cv.circle(img1, tuple(pt1), 5, color, 5)
    return img1
```
