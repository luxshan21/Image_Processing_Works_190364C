```python
def exe_time_eval(n):
  start_time = time.time()
  dummy = []
  for x in range(n):
    dummy.append(x)
  return( time.time() - start_time)
times = []
n=1000
for x in range(n):
  t = 0.0
  for y in range(10):
    t+=exe_time_eval(x)
  times.append((t/10))
print(times)
import matplotlib.pyplot as plt
elements = range(n)
plt.xlabel('List Length')
plt.ylabel('Time Complexity')
plt.plot(elements, times, label ='loop')
plt.grid()
plt.legend()
plt.show()
```
