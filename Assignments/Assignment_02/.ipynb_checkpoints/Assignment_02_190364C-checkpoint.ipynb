```python
# Question 1 (a)
import numpy as np
from numpy.linalg import inv
def Random_Points_Generator(datas, k): #This funcion will return k number of  random points from the data
    sample_datas = []
    sum = 0
    while sum < k:
        ind = np.random.randint(len(datas))
        x = datas[ind][0]
        y = datas[ind][1]
        if (x, y) not in sample_datas:
            sample_datas.append((x, y))
            sum += 1
    return sample_datas
def Parameters_Of_circle(sample_datas): #This funcion will return the parameters of circle from sample datas
    point_1 = sample_datas[0]
    point_2 = sample_datas[1]
    point_3 = sample_datas[2]
    A = np.array([[point_2[0] - point_1[0], point_2[1] - point_1[1]], [point_3[0] - point_2[0], point_3[1] - point_2[1]]]) 
    B = np.array([[point_2[0]**2 - point_1[0]**2 + point_2[1]**2 - point_1[1]**2], [point_3[0]**2 - point_2[0]**2 + point_3[1]**2 - point_2[1]**2]])
    inv_A = inv(A)
    c, d = np.dot(inv_A, B)/2
    c, d = c[0], d[0]
    r = np.sqrt((c-point_1[0])**2 + (d-point_1[1])**2)
    return (c, d, r)
def Inliers_And_Distance(circle, datas, threshold): #This funcion will return total distance and inlier
    x_data = datas[:, 0]
    y_data = datas[:, 1]
    c = circle[0]
    d = circle[1]
    r = circle[2]
    inliers = []
    total_distance = 0
    for i in range(len(x_data)):
        distance = np.sqrt((x_data[i] - c)**2 + (y_data[i] - d)**2)

        if abs(distance - r) <= threshold:
            total_distance += abs(distance - r)
            inliers.append([x_data[i], y_data[i]])  
    return inliers, total_distance
```

```python
# Question 1 (b)

import numpy as np
from scipy.optimize import minimize
from scipy import linalg
import matplotlib.pyplot as plt
# np.random.seed(0)
N = 100
half_n = N//2
r = 10
s = r /16
t = np.random.uniform(0 , 2*np.pi , half_n)
n = s*np.random.randn(half_n)
x , y = (r + n)*np.cos(t),(r + n)*np.sin(t)
X_circ = np.hstack((x.reshape(half_n , 1), y.reshape(half_n ,1)))
m, b = -1, 2
x = np.linspace(-12, 12 , half_n)
y = m*x + b + s*np.random.randn(half_n)
X_line = np.hstack ( ( x.reshape(half_n,1), y.reshape(half_n ,1)))
X = np.vstack((X_circ,X_line))
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(X[:,0], X[:, 1], 'o', label='Outliers')
ax.grid(True)
x_data = X[:, 0]
y_data = X[:, 1]
iterations = 35
inlier_threshold = r/8
inlier_count_threshold = 50
best_model = [0, 0, 999999, []]

for i in range(iterations):
    sample = Random_Points_Generator(X, 3)# obtain three points from data
    initial_circle = Parameters_Of_circle(sample)
    if abs(r - initial_circle[2]) > 1.5*r: #remove larger circles
        continue
    inliers, toal_distance = Inliers_And_Distance(initial_circle, X, inlier_threshold)
    if len(inliers) < inlier_count_threshold:# start from beginning when inlier count is less 
        continue
    inlier_sample =  Random_Points_Generator(inliers, 3)
    circle = Parameters_Of_circle(inlier_sample) # find circle parameters from three inlier points
    inliers, toal_distance = Inliers_And_Distance(circle, X, inlier_threshold) #find new inliers and total distance
    if len(inliers) < inlier_count_threshold:
        continue
    if len(best_model[-1]) < len(inliers):
        best_model = [circle, initial_circle, sample, toal_distance, inliers]
    elif len(best_model[-1]) == len(inliers):
        if best_model[3] > toal_distance:
            best_model = [circle, initial_circle, sample, toal_distance, inliers]
if len(best_model[-1]) == 0:
    print("best model found Not Find")
else:
    inliers = np.array(best_model[-1])
    ax.plot(inliers[:, 0], inliers[:, 1], 'o', label='Inliers')
    best_samples = np.array(best_model[2])
    ax.plot(best_samples[:,0], best_samples[:, 1], 'o', label='Best samples')
    a, b, r = best_model[0][0], best_model[0][1], best_model[0][2]
    circle = circle((a, b), radius=r, color='r', fc='y', fill=False, label='RANSAC')
    ax.add_patch(circle)
    a, b, r = best_model[1][0], best_model[1][1], best_model[1][2]
    init_circle = circle((a, b), radius=r, color='b', fc='y', fill=False, label='Circle estimated from best sample')
    ax.add_patch(init_circle) 
ax.legend();
```

```python
# Question 3 (b)
def Homo(src_img, dst_img): # Get the Homography
    sift = cv.SIFT_create()
    p1, des1 = sift.detectAndCompute(src_img, None)#  keypoints and descriptors with SIFT
    p2, des2 = sift.detectAndCompute(dst_img, None)
    FLANN_INDEX_KDTREE = 1
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5) # parameters of FLANN
    search_params = dict(checks = 50)  
    flann = cv.FlannBasedMatcher(index_params, search_params)
    matches = flann.knnMatch(des1, des2, k = 2)
    good_points = []
    for m,n in matches:
        if m.distance < 0.7*n.distance: # store all the good matches points  
            good_points.append(m)
    src_pts = np.float32([p1[m.queryIdx].pt for m in good_points])
    dst_pts = np.float32([p2[m.trainIdx].pt for m in good_points])
    Homography, mask = cv.findHomography(src_pts.reshape(-1,1,2), dst_pts.reshape(-1,1,2), cv.RANSAC, 0.95)
    return Homography
Homography = np.identity(3)
for i in range(4):
    source = cv.imread('img'+str(i + 1)+'.ppm', cv.IMREAD_GRAYSCALE)
    destination = cv.imread('img'+str(i + 2)+'.ppm', cv.IMREAD_GRAYSCALE)
    Homography = np.matmul(Homo(source,destination), Homography)
print(Homography)
```
